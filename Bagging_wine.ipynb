### Outliers handling

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

IQR = Q3-Q1

min_limit = Q1-1.5*IQR
max_limit = Q3+1.5*IQR

outliers_count = ((data > max_limit) | (data < min_limit)).sum()

outliers_percentage = (outliers_count/len(data))*100

print('\n Sum of Outliers:\n',outliers_count)
print('\n Percentage of Outliers:\n',outliers_percentage)
```

```python
plt.figure(figsize=(20,20),facecolor='white')
plotnumber=1

for column in data.drop('wine',axis=1):
        if plotnumber<9:
            ax=plt.subplot(4,2,plotnumber)
            sns.boxplot(x=data[column])
            plt.xlabel(column)
        plotnumber+=1
plt.tight_layout()
```

```python
## m_acid

from scipy import stats

Q1 = np.percentile(data.m_acid,25)
Q3 = np.percentile(data.m_acid,75)

IQR = stats.iqr(data['m_acid'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(data['m_acid'] > max_limit) | (data['m_acid'] < min_limit),'m_acid'] = np.median(data['m_acid'])
```

```python
## ash

Q1 = data.ash.quantile(0.25)
Q3 = data.ash.quantile(0.75)
IQR = stats.iqr(data['ash'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(data['ash'] > max_limit) | (data['ash'] < min_limit),'ash'] = np.mean(data['ash'])
```

```python
## AOA

mean = np.mean(data['AOA'])

std = np.std(data['AOA'])

min_value = mean-(3*std)
max_value = mean+(3*std)

data.loc[(data['AOA'] > max_value) | (data['AOA'] < min_value),'AOA'] = np.mean(data['AOA'])
```

```python
## Magnesium

Q1 = data.Magnesium.quantile(0.25)
Q3 = data.Magnesium.quantile(0.75)
IQR = stats.iqr(data['Magnesium'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(data['Magnesium'] > max_limit) | (data['Magnesium'] < min_limit),'Magnesium'] = np.median(data['Magnesium'])
```

```python
## pro_cyanin

Q1 = data.pro_cyanin.quantile(0.25)
Q3 = data.pro_cyanin.quantile(0.75)
IQR = stats.iqr(data['pro_cyanin'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(data['pro_cyanin'] > max_limit) | (data['pro_cyanin'] < min_limit),'pro_cyanin'] = np.median(data['pro_cyanin'])
```

```python
## color_intensity

Q1 = data.color_intensity.quantile(0.25)
Q3 = data.color_intensity.quantile(0.75)
IQR = stats.iqr(data['color_intensity'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(data['color_intensity'] > max_limit) | (data['color_intensity'] < min_limit),'color_intensity'] = np.median(data['color_intensity'])
```

```python
## hue

Q1 = data.hue.quantile(0.25)
Q3 = data.hue.quantile(0.75)
IQR = stats.iqr(data['hue'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(data['hue'] > max_limit) | (data['hue'] < min_limit),'hue'] = np.median(data['hue'])
```

### scaling

```python
# splitting the data

x = data.drop('wine',axis=1)
y = data.wine
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42)
```

```python
## model implementation

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

model = SVC(kernel='linear')

model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

## Bagging

```python
from sklearn.ensemble import BaggingClassifier

model_bag = BaggingClassifier(base_estimator=model,n_estimators=20)

model_bag.fit(x_train,y_train)
```

```python
y_pred_bag = model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_bag)
```
